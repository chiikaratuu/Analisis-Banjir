#Data Cleaning

```python
# Mengisi kekosongan data pada kolom manggarai dengan nilai rata-rata dari baris yang sesuai
group_cols = ['Katulampa', 'Pos Depok', 'Istiqlal', 'Jembatan Merah', 'Flusing Ancol', 'Marina Ancol', 'Status Banjir', 'Bulan', 'Hari', 'Jam']
for col in group_cols:
    df['Manggarai'].fillna(df.groupby([col, 'Status Banjir'])['Manggarai'].transform('mean'), inplace=True)
print(df)
```

```python
df = df.drop('Tahun', axis=1)
df = df.drop('Menit', axis=1)
df = df.drop('Detik', axis=1)
```

```python
# Memisahkan variabel independen (X) dan dependen (y)
X = df.drop(columns=['Status Banjir'])  # Variabel prediktor, kecuali 'Status Banjir'
y = df['Status Banjir']  # Variabel target
```

#Balancing Data

```python
from imblearn.over_sampling import SMOTE
# Inisialisasi SMOTE
smote = SMOTE(random_state=42)

# Fit dan transform data
X_resampled, y_resampled = smote.fit_resample(X, y)

# Memeriksa jumlah data setelah SMOTE
print("Jumlah data sebelum SMOTE:", y.value_counts())
print("Jumlah data setelah SMote:", pd.Series(y_resampled).value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data sebelum SMOTE
before_smote_counts = y.value_counts()

# Data setelah SMOTE
after_smote_counts = pd.Series(y_resampled).value_counts()

# Persiapkan data untuk bar chart
data = pd.DataFrame({'Class': before_smote_counts.index, 'Before SMOTE': before_smote_counts.values, 'After SMOTE': after_smote_counts.values})

# Plot bar chart
data.plot(x='Class', kind='bar')
plt.xlabel('Class')
plt.ylabel('Jumlah Data')
plt.title('Perbandingan Jumlah Data Sebelum dan Sesudah SMOTE')
plt.show()
```

#Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_normalized = scaler.fit_transform(X_resampled)
```

```python
import pandas as pd

# Konversi X_normalized menjadi DataFrame
X_normalized_df = pd.DataFrame(X_normalized, columns=X.columns)

X_normalized_df.head()
```

#Splitting Data

```python
from sklearn.model_selection import train_test_split

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y_resampled, test_size=0.2, random_state=42)
```

```python
import pandas as pd

# Konversi array ke dalam DataFrame (opsional, jika belum dalam format DataFrame)
X_train_df = pd.DataFrame(X_train)
X_test_df = pd.DataFrame(X_test)

# Jumlah keseluruhan data training dan test
total_train = len(X_train_df)
total_test = len(X_test_df)

print("Jumlah keseluruhan data training:", total_train)
print("Jumlah keseluruhan data test:", total_test)
```

#Modelling

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


# Inisialisasi model Decision Tree
model_dt = DecisionTreeClassifier()

# Melatih model Decision Tree
model_dt.fit(X_train, y_train)

# Menguji model
y_pred_dt = model_dt.predict(X_test)

# Menghitung akurasi
accuracy_dt = accuracy_score(y_test, y_pred_dt)

print("Akurasi model Decision Tree:", round(accuracy_dt, 3))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Inisialisasi model Random Forest
model_rf = RandomForestClassifier()

# Melatih model Random Forest
model_rf.fit(X_train, y_train)

# Menguji model
y_pred_rf = model_rf.predict(X_test)

# Menghitung akurasi
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print("Akurasi model Random Forest:", round(accuracy_rf, 3))
```

#Classification Report

```python
from sklearn.metrics import classification_report

# Membuat classification report untuk model Decision Tree
report_dt = classification_report(y_test, y_pred_dt)

# Membuat classification report untuk model Random Forest
report_rf = classification_report(y_test, y_pred_rf)

# Menampilkan classification report untuk model Decision Tree
print("Classification Report - Decision Tree:\n", report_dt)

# Menampilkan classification report untuk model Random Forest
print("Classification Report - Random Forest:\n", report_rf)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Menghitung confusion matrix untuk model Decision Tree
cm_dt = confusion_matrix(y_test, y_pred_dt)

# Visualisasi confusion matrix untuk model Decision Tree
plt.figure(figsize=(8, 6))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Decision Tree')
plt.show()

# Menghitung confusion matrix untuk model Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf)

# Visualisasi confusion matrix untuk model Random Forest
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest')
plt.show()
```
